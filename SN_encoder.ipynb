#### code snippet A1

```python
import os
import json

# Serial number data to be written to the JSON file
EMIT_SN = {
    "VF": 15,   
    "A" : 20, 
    "AS": 100,
    "ts": "TSN",
    "type": "P-type"
}

# File path for the JSON file
json_file_path = "EMIT_SN.json"

# writing data to the JSON file
with open(json_file_path, 'w') as json_file:
    json.dump(EMIT_SN, json_file, indent=4)
    
print(f"JSON file '{json_file_path}' has been created")
```

#### code snippet A2

```python
# Reading SN data from JSON file
with open(json_file_path, 'r') as json_file:
    data = json.load(json_file)
    
# Constructing the string
folder_name = f"EMIT-VF{data['VF']}A{data['A']}AS{data['AS']}{data['ts']}-{data['type']}"

# Creating the folder
os.makedirs(folder_name, exist_ok=True)

print(f"Folder '{folder_name}' has been created.")
```

#### Code snippet A3 and A4 for multiple SFID-tags for audio folders

```python
import os
import json

# Serial number data to be written to the JSON file
EMIT_SN = [
    {
        "VF": 15,
        "A": 20,
        "AS": 100,
        "ts": "TSN",
        "type": "P-type"
    },
    {
        "VF": 15,
        "A": 20,
        "AS": 85,
        "ts": "PS06",
        "type": "GP-type"
    },
    {
        "VF": 20,
        "A": 20,
        "AS": 100,
        "ts": "G06",
        "type": "Gyroid"
    }
]

# File path for the JSON file
json_file_path = "EMIT_SN.json"

# Writing data to the JSON file
with open(json_file_path, 'w') as json_file:
    json.dump(EMIT_SN, json_file, indent=4)

print(f"JSON file '{json_file_path}' has been created")

# Reading SN data from JSON file
with open(json_file_path, 'r') as json_file:
    data = json.load(json_file)

# Creating folders based on the data
for item in data:
    folder_name = f"EMIT-VF{item['VF']}A{item['A']}AS{item['AS']}{item['ts']}-{item['type']}"
    os.makedirs(folder_name, exist_ok=True)
    print(f"Folder '{folder_name}' has been created.")
```
